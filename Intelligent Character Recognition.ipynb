# Intelligent Character Recognition
(AKA ICR)

See:
- https://en.wikipedia.org/wiki/Intelligent_character_recognition

This is not the same as OCR.

> When we talk about OCR, we are really talking about converting data from an image that was created by a machine, whether it be a document created by an office application, and even an old document typewritten
>
> For ICR, pattern-matching goes to a different level. You cannot really have success deploying a solution requiring ICR without additional information to aid in the recognition process.
>
> https://www.parascript.com/blog/difference-ocr-icr/

# Let's explore the available options on the market

## queXF
An open source, web based paper form verification and data entry system

> The process of ICR is broken in to the following 7 steps:
> - Character isolation
> - Noise reduction
> - Boundary removal
> - Normalising
> - Thinning
> - Feature extraction
> - Training or Recognition
>
> https://quexf.acspri.org.au/intelligent-character-recognition

It can:
- Automatically detects locations of boxes on the form (no need to manually overlay boxes)


## NIST FORM-BASED HANDPRINT RECOGNITION SYSTEM (RELEASE 2.2)
A public domain document processing system was developed by the National Institute of Standards and Technology (NIST) in 1994. The system is a standard reference form-based handprint recognition system for evaluating optical character recognition (OCR), and it is intended to provide a baseline of performance on an open application. 

NIST developed a new release in 2003 for internet distribution of its standard reference form-based handprint recognition system for evaluating optical character recognition (OCR). Release 2.2 incorporates all bug fixes generated for previous versions.

https://www.nist.gov/services-resources/software/public-domain-ocr

For source code and training data:
- https://github.com/brainopener/nist-ocr

According to the documentation:
- To process a new type of form, trainreg must be run on a prototypical form and the output coordinate stored.
